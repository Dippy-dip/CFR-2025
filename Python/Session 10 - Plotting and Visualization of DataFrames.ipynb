#### Example 4: Histogram

Histograms are useful for displaying the distribution of a dataset.


```python
# Generate our data
#********
# Step 1: Generate 1000 random numbers 
data = np.random.randn(1000)
data[:11]
# Step 2 : create data frame with the data
df = pd.DataFrame(data, columns =['Values'])
df.head(10)

# Step 3: save the DataFrame to a csv file (can be used in Excel)
# creata a variable for a csv file name, so...
name_csv_file = 'data_seccion10_example4.csv'
df.to_csv(name_csv_file, index=False)

# Step 4: Bring in the data from the csv
df_from_csv = pd.read_csv(name_csv_file)
df_from_csv.head(10)

# Create plot
# ********
plt.hist(df_from_csv['Values'], bins=30, edgecolor='red')
plt.title('Histogram of Values')
plt.xlabel('Values')
plt.ylabel('Frequency')
```

## Part 3: Advanced Customization and Subplots (30 minutes)


#### Example 6: Creating Subplots

Subplots allow you to create multiple plots in a single figure.

```python
# Sample Data
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
y2 = [1, 4, 6, 8, 10]

#Create a Data Frame form sample data
df = pd.DataFrame({'X':x, 'Y1':y1, 'Y2':y2})
print(df)

# Create graphs
plt.subplot(3,2,1)
plt.plot(df['X'], df['Y1'], color='red')

plt.subplot(3,2,2)
plt.plot(df['X'], df['Y1'], color='yellow')

plt.subplot(3,2,3)
plt.plot(df['X'], df['Y1'], color='blue')

plt.subplot(3,2,4)
plt.plot(df['X'], df['Y1'], color='green')

plt.subplot(3,2,5)
plt.plot(df['X'], df['Y1'], color='orange')

plt.subplot(3,2,6)
plt.plot(df['X'], df['Y1'], color='purple')
```
